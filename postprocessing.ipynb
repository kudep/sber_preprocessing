```python
import os, io
bot_answers = []
base_dir = '/home/kuznetsov/tmp/thread33/results'
# base_file_name = 'd_train'
# base_file_name = 'd_train_one_fifth_low'
base_file_name = 'd_train_one_fifth_mid'


corfilename = os.path.join(base_dir, base_file_name + '.cor')
with io.open(corfilename, encoding='utf-8') as file:
    contexts = file.readlines()
    
print('Number of examples in cor file: ', len(contexts))

manfilename = os.path.join(base_dir, base_file_name + '.man')
with io.open(manfilename, encoding='utf-8') as file:
    human_answers = file.readlines()
    
print('Number of examples in man file: ', len(human_answers))

step = 457000
num_of_files = 10

bot_answers_set = [set() for _ in range(len(human_answers))]
for k in range(num_of_files):
    filename = os.path.join(base_dir, base_file_name + '-' + str(step) + '-' + str(k) + '.nmt')
    with io.open(filename, encoding='utf-8') as file:
        for ans_set, line in zip(bot_answers_set, file.readlines()):
            ans_set.add(line)
```

```python
new_bot_ans = []
new_hum_ans = []
new_contexts = []
for bot_set_ans, human_ans, context in zip(bot_answers_set, human_answers, contexts):
    if human_ans in bot_set_ans: 
        bot_set_ans.remove(human_ans)
    if len(bot_set_ans) > 0:
        new_bot_ans.append(list(bot_set_ans))
        new_hum_ans.append(human_ans)
        new_contexts.append(context)
```

```python
base_dir = '/home/kuznetsov/tmp/thread33/postprep'
corfilename = os.path.join(base_dir, base_file_name + '.cor')
with open(corfilename, 'wt') as file:
    file.writelines(new_contexts)

manfilename = os.path.join(base_dir, base_file_name + '.man')
with open(manfilename, 'wt') as file:
    file.writelines(new_hum_ans)

for k in range(num_of_files):
    filename = os.path.join(base_dir, base_file_name + '-' + str(step) + '-' + str(k) + '.nmt')
    with open(filename, 'wt') as file:
        for bot_ans in new_bot_ans:
            file.write(bot_ans[k%len(bot_ans)])
```
